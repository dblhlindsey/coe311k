```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
h_values = [0.1, 0.01, 0.001, .00001, .0000001] # Different step sizes
t_start = 0 # Start time
t_end = 30 # End time
y_0 = 0.0 # Initial displacement

# Derivative Function
def df(t,y):
    return -1000*y + 3000 - 2000*np.exp(-1*t)

# Exact Function
def f(t):
    return 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)

# Function to apply Euler's Method for given step size
def euler_forward_method(h, t_start, t_end, y_0):
    t_values = np.arange(t_start, t_end + h, h)
    y_values = np.zeros(len(t_values))
    
    # Set initial conditions
    y_values[0] = y_0
    
    for i in range(1, len(t_values)):
        y_n = y_values[i - 1]
        # Update equations for Euler's Method
        y_values[i] = y_n + h * df(t_values[i-1], y_values[i-1])
    return t_values, y_values
```

```python
for h in h_values:
    t, y = euler_forward_method(h, t_start, t_end, y_0)
    plt.plot(t, y, label=f'Euler\'s Forward (h={h})')

t_exact = np.linspace(t_start, 30, 1000)
plt.plot(t_exact, f(t_exact), label='Exact Solution')
    
plt.xlabel('Time (t)')
plt.ylabel('Displacement (y)')
plt.title('Stability Analysis of Euler\'s Forward')
plt.legend()
plt.grid(True)
plt.show()
```
