```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
def montepifunc(numSamples):
    counter = 0
    for n in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2+y**2)<=1:
            counter += 1
    pi = 4*counter/numSamples
    return pi

absolute = []
relative = []

for i in range(0,100):
    x = montepifunc(1000)
    absolute.append(np.abs(np.pi - x))
    relative.append(np.abs(np.pi - x)/x)
    
x = np.linspace(1,100,100)
    
plt.plot(x, absolute, label = "absolute error")
plt.plot(x, relative, label = "relative error")
plt.xlabel("iteration")
plt.ylabel("error")
plt.legend()
plt.show()
```
