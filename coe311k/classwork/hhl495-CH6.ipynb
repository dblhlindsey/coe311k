```python
#a)
A = np.array([[10, -1, 2, 0, 0],
[-1, 11, -1, 3, 0],
[2, -1, 10, -1, 0],
[0, 3, -1, 8, -2],
[0, 0, 0, -2, 9]])
b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b, dtype = 'float')

start_time = time.perf_counter()
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
end_time = time.perf_counter()
print(f"Jacobi Matrix 1 Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Run Time: {(end_time - start_time)*1e3} ms")

print("")

start_time = time.perf_counter()
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)
end_time = time.perf_counter()
print(f"Gauss Seidel 1 Matrix Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Run Time: {(end_time - start_time)*1e3} ms")

#Both methods converge for this matrix. The Gauss Seidel matrix converges significantly faster, nearly twice as fast.
#This is because the spectral radius for the diagonal of the Gauss Seidel method smaller, allowing it to update more
#accurately, as well as update in place. The Jacobi method took 18 iterations and the Gauss-Seidel method took 10 iterations.
```
