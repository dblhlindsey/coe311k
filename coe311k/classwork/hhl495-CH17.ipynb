```python
import numpy as np
import matplotlib.pyplot as plt

def eulerapproxh(h,k):
    # Define the parameters
    # Constant in the equation (e.g., spring constant)
    # Step size
    x_start = 0 # Start of the interval
    x_end = 10 # End of the interval
    y_0 = 1.0 # Initial condition for y
    v_0 = 0.0 # Initial condition for dy/dx

    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
    
    # Plot the results
    plt.plot(x_values, y_values, label='Position y(x) for h = ' + str(h))
    plt.plot(x_values, v_values, label='Velocity v(x) for h = ' + str(h),
    linestyle='--')
    
def eulerapproxk(h,k):
    # Define the parameters
    # Constant in the equation (e.g., spring constant)
    # Step size
    x_start = 0 # Start of the interval
    x_end = 10 # End of the interval
    y_0 = 1.0 # Initial condition for y
    v_0 = 0.0 # Initial condition for dy/dx

    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
    
    # Plot the results
    plt.plot(x_values, y_values, label='Position y(x) for k = ' + str(k))
    plt.plot(x_values, v_values, label='Velocity v(x) for k = ' + str(k),
    linestyle='--')
```

```python
plt.figure(figsize=(10, 5))
eulerapproxh(.01,1)
eulerapproxh(.02,1)
eulerapproxh(.03,1)
eulerapproxh(.05,1)
eulerapproxh(.1,1)
eulerapproxh(.2,1)
eulerapproxh(.3,1)
eulerapproxh(.5,1)
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend(fontsize=7)
plt.grid(True)
plt.show()
```

It seems to become unstable at h = 0.5 where the position function seems to diverge. It does become more unstable as h increases because lambda grows with h. Whenever the k value increases, the amplitude of the velocity function increases. However, as k increases so does the frequency of the v(x) function, so the amplitude of the y(t) function doesn't necessarily increase. Euler's method cannot solve this with a larger h because lambda becomes larger and positive where it becomes unstable.
