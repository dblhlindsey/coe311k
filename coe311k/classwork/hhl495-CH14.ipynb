```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)
    
# Params
a = 0
b = 2*np.pi
N = 200

# Init x
x = np.linspace(a, b, N)

# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_cdx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_fdx = (f(x + h) - f(x)) / (h)

# Backward Difference Approximation
df_bdx = (f(x) - f(x - h)) / (h)

# True Deriv
df_dx = np.cos(x)

# Plot the approximation and the true derivative
## Do a canvas of 4 subplots or 4 graphs on one plot comparing cos(x), Forward Difference, Backwards Difference, and Central Difference Methods

plt.plot(x, df_dx, label="cos(x)")
plt.plot(x, df_cdx, linestyle='--', label="Central")
plt.plot(x, df_fdx, linestyle='--', label="Forward")
plt.plot(x, df_bdx, linestyle='--', label="Backward")

plt.legend()
plt.title("Derivative Approximations of sin(x)")
plt.show()
```
