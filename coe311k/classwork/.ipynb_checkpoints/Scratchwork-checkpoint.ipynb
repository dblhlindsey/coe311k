```python
import math
import random
def montepifunc(numSamples):
    counter = 0
    for n in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2+y**2)<=1:
            counter += 1
    pi = 4*counter/numSamples
    return pi
print(montepifunc(99999999))
```
