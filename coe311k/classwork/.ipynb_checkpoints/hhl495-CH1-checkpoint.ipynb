```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f1(x,y):
    return (np.arcsin(x)+np.arcsin(y))

def f2(x,y):
    return (np.sin(x)+np.cos(x))

def f3(x,y):
    return (x*x+y*y)

def f4(x,y):
    return (x+y)

x = np.linspace(-1, 1, 2000)
y = np.linspace(-1, 1, 2000)
X, Y = np.meshgrid(x, y)
A = f1(X, Y)
B = f2(X, Y)
C = f3(X, Y)
D = f4(X, Y)

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
ax1.contour3D(X, Y, A, 50, cmap='hot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z');
ax1.view_init(45, 0)
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax2.contour3D(X, Y, B, 50, cmap='hot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z');
ax2.view_init(45, 0)
ax3 = fig.add_subplot(2, 2, 3, projection='3d')
ax3.contour3D(X, Y, C, 50, cmap='hot')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z');
ax3.view_init(45, 0)
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
ax4.contour3D(X, Y, D, 50, cmap='hot')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_zlabel('z');
ax4.view_init(45, 0)
```
