```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time

def matrix_multiplication(a, b):
    n = len(a)
    c = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                c[i][j] += a[i][k] * b[k][j]
                
    return c

runtimesnaive = []
runtimespy = []
x = np.arange(1,9)

for i in range(0, 8):
    n = 3 * (2**i)
    a, b = np.zeros((n,n)), np.zeros((n,n))
    
    for j in range (0, n):
        for k in range (0, n):
            a[i][j] = random.uniform(0,100)
            
    for w in range (0,n):
        for y in range(0,n):
            b[i][j] = random.uniform(0,100)
            
    print(n, "x", n, " ( Matrix ", i,")")
    
    t1= time.perf_counter()
    matrix_multiplication(a, b)
    t2= time.perf_counter()
    print("Naive time: ", t2-t1)
    runtimesnaive.append(t2-t1)
    
    t1= time.perf_counter()
    a @ b
    t2 = time.perf_counter()
    print("Python time: ", t2-t1)
    runtimespy.append(t2-t1)
    print("\n")
    

plt.plot(x , runtimesnaive, label="Custom Matrix Multiplication")
plt.plot(x, runtimespy, label="Python Matrix Multiplication")
plt.xlabel("Matrix")
plt.ylabel("Time (s)")
plt.title("Matrix vs Runtime for Different Matrix Multiplication")
plt.legend()
plt.show()
```
