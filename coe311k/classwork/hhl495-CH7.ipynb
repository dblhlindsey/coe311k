```python
import numpy as np
import time

def diagonalDominance(array):
    strength = "Strictly Dominant"
    for i in range(len(array)):
        sum = 0
        for j in range(len(array)):
            if (j != i):
                sum += abs(array[i,j])
        if (abs(array[i,i]) < sum):
            return "Not Diagonally Dominant"
        if  (array[i,i] == sum):
            strength = "Weakly Dominant"
    return strength

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

tol = 1e-6
max_iterations = 300000
```

```python
#a)
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(A)


print("Part A")
print("\n")

start_time = time.perf_counter()
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
end_time = time.perf_counter()
print(f"Jacobi Solution: {solution}")
print(f"Jacobi Iterations: {iterations}")
print(f"Jacobi Run-time: {(end_time - start_time)*1e3} ms")

print("\n")

start_time = time.perf_counter()
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)
end_time = time.perf_counter()
print(f"Gauss Seidel Solution: {solution}")
print(f"Gauss Seidel Iterations: {iterations}")
print(f"Gauss Seidel Run-time: {(end_time - start_time)*1e3} ms")
print("Diagonal Dominance:")
print(diagonalDominance(A))
```
