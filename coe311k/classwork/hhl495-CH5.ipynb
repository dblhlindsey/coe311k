```python
import numpy as np
rng = np.random.default_rng()

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

A = np.array([[10, 1, 2, 3, 1],
[1, 12, 1, 2, 3],
[2, 1, 14, 1, 2],
[3, 2, 1, 15, 1],
[1, 3, 2, 1, 16]])
B = np.array([[2, 4, 5, 3, 6],
[3, 1, 4, 5, 2],
[6, 2, 3, 4, 1],
[5, 3, 6, 2, 4],
[4, 6, 2, 5, 3]])

C = 20 * rng.random(5)
guess = np.zeros(5)
max_iterations = 1000
tolerance = 10**-6
```

```python
solution, iterations = jacobi(A, C, guess, tolerance, max_iterations)
print(f"Jacobi Matrix A Solution: {solution}")
print(f"Iterations: {iterations}")
solution, iterations = jacobi(B, C, guess, tolerance, max_iterations)
print(f"Jacobi Matrix B Solution: {solution}")
print(f"Iterations: {iterations}")
```

```python
solution, iterations = gauss_seidel(A, C, guess, tolerance, max_iterations)
print(f"Gauss Seidel A Matrix Solution: {solution}")
print(f"Iterations: {iterations}")
solution, iterations = gauss_seidel(B, C, guess, tolerance, max_iterations)
print(f"Gauss Seidel Matrix B Solution: {solution}")
print(f"Iterations: {iterations}")
```

```python
#Matrix B does not converge because it is not diagonally dominant
#Because it is not diagonally dominant these solving methods amplify the error.
#For Matrix A, Gauss Seidel reaches the tolerance faster and with less time due to its ability to update in place
#and not use a second matrix as well as taking in account the sum of the new in addition to the sum of the old 
#to iterate in less calculations.
```
