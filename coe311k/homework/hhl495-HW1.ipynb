```python
import math
import random
import numpy as np
def montepifunc(numSamples):
    counter = 0
    for n in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2+y**2)<=1:
            counter += 1
    pi = 4*counter/numSamples
    return pi
```

```python
#I'm starting the samples at 20, because 990 isn't divisible by 20 despite the assignment asking to start at 10

for i in range(1,50):
    x = montepifunc(i * 20)
    print("For", i*20, "samples...")
    print("The estimated pi was", x)
    print("The difference was %.15f" % np.abs(np.pi - x))
    print("")
```
