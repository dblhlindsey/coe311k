```python
import math
import numpy as np
import matplotlib.pyplot as plt

def sinapprox(numTerms, x):
    approx = 0
    for i in range (0, numTerms):
        approx = approx + (x**(2*i+1))*((-1)**i)*(1/(np.math.factorial((2*i)+1)))
    return approx
```

```python
errorOneTerm = []
errorTwoTerm = []
errorThreeTerm = []
errorFourTerm = []
i=1
for j in range (0, 100):
    i = j/10
    errorOneTerm.append(abs(sinapprox(1,i) - np.sin(i)))
    errorTwoTerm.append(abs(sinapprox(2,i) - np.sin(i)))
    errorThreeTerm.append(abs(sinapprox(3,i) - np.sin(i)))
    errorFourTerm.append(abs(sinapprox(4,i) - np.sin(i)))
```

```python
x = np.arange(0, 10, 0.1)
plt.plot(x,errorOneTerm,label = "one term")
plt.plot(x,errorTwoTerm,label = "two term")
plt.plot(x,errorThreeTerm,label = "three term")
plt.plot(x,errorFourTerm,label = "four term")
plt.xlabel("X Value")
plt.ylabel("Absolute Error")
plt.legend()
plt.show()
```

```python
#The truncation error increases as you add more terms for x values away from zero, and decreases for x values near zero
#This is because when you have terms like (x^5)/5! it will increase accuracy at small x values, but the value will blow up
#when farther from x, so truncation error does not behave uniformly for all values of x
```
